```python
class CPU:
    def __init__(self, core, thread, gen) -> None:
        self.core = core
        self.gen = gen
        self.thread = thread

    def show_cpu(self):
        return f'Cores: {self.core}, Threads: {self.thread}, Generation: {self.gen}'.upper()


class RAM:
    def __init__(self, type, memory, speed) -> None:
        self.type = type
        self.memory = memory
        self.speed = speed

    def show_ram(self):
        return f'Type: {self.type}, Memory: {self.memory}, Speed: {self.speed}'.upper()


class StorageDevices:
    def __init__(self, type, memory, speed) -> None:
        self.type = type
        self.memory = memory
        self.speed = speed

    def show_storage(self):
        return f'Type: {self.type}, Memory: {self.memory}, Speed: {self.speed}'.upper()


class ComputerSystem:
    def __init__(self, name) -> None:
        self.name = name
        self.ram_component = RAM('DDR4', '16GB', '337MHz')
        self.storage_component = StorageDevices('SSD', '256GB', '9800MB/s')
        self.cpu_component = CPU(10, 11, 'i7-11th-gen')

    def get_specs(self):
        cpu_info = self.cpu_component.show_cpu()
        ram_info = self.ram_component.show_ram()
        storage_info = self.storage_component.show_storage()
        return cpu_info, ram_info, storage_info

    def print_specs(self):
        cpu_info, ram_info, storage_info = self.get_specs()
        print(f'''
        YOUR COMPUTER NAME: {self.name}
        YOUR COMPUTER SPECS:
        {cpu_info}
        {ram_info}
        {storage_info}
        ''')


my_computer = ComputerSystem("My Awesome Computer")
my_computer.print_specs()
```

```python
# 10. Create a Python class for a car, using composition to represent components 
#like the engine, wheels, and transmission.
class engine:
    def __init__(self,type,cc) -> None:
        self.__type=type
        self.__cc=cc

    def show_en(self):
        return f'type : {self.__type} CC : {self.__cc}'
    
class wheel:
    def __init__(self,dia,type):
        self.__dia=dia
        self.__type=type

    def show_w(self):
        return f'diameter : {self.__dia} type :{self.__type}'
    
class transmission:
    def __init__(self,type,further_type) -> None:
        self.__type=type
        self.__fur_type=further_type

    def show_t(self):
        return f"type {self.__type} and specs {self.__fur_type}"

class car:
    def __init__(self,model,color) -> None:
        self.__model=model
        self.__color=color
        self.__wheel_c=wheel(10.5,'off-roading')
        self.__transmisson_c=transmission('manual','5-gear')
        self.__engine_c=engine('euro V','1299cc')


    def __show_specs(self):
        return str("\n"+self.__wheel_c.show_w()+"\n"+self.__transmisson_c.show_t()+"\n"+self.__engine_c.show_en())
    
    def all_specs(self):
        print(f' car model:{self.__model} car color {self.__color}')
        print(f' ALL SPECS \n {self.__show_specs()}')

c1=car(model='honda70',color='BLACK')
c1.all_specs()
```

```python
class Appliance:
    def __init__(self, name: str, brand: str):
        self.name = name
        self.brand = brand

    def __str__(self):
        return f"Appliance: {self.name} (Brand: {self.brand})"

class Furniture:
    def __init__(self, name: str, material: str):
        self.name = name
        self.material = material

    def __str__(self):
        return f"Furniture: {self.name} (Material: {self.material})"

class Room:
    def __init__(self, name: str):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture: Furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance: Appliance):
        self.appliances.append(appliance)

    def __str__(self):
        furniture_str = "\n  ".join(str(furniture) for furniture in self.furniture)
        appliances_str = "\n  ".join(str(appliance) for appliance in self.appliances)
        return f"Room: {self.name}\nFurniture:\n  {furniture_str}\nAppliances:\n  {appliances_str}"

class House:
    def __init__(self, address: str):
        self.address = address
        self.rooms = []

    def add_room(self, room: Room):
        self.rooms.append(room)

    def __str__(self):
        rooms_str = "\n".join(str(room) for room in self.rooms)
        return f"House at {self.address}\nRooms:\n{rooms_str}"

# Example usage

sofa = Furniture("Sofa", "Leather")
table = Furniture("Dining Table", "Wood")
fridge = Appliance("Refrigerator", "Samsung")
oven = Appliance("Oven", "LG")

living_room = Room("Living Room")
living_room.add_furniture(sofa)
living_room.add_appliance(fridge)

kitchen = Room("Kitchen")
kitchen.add_furniture(table)
kitchen.add_appliance(oven)

my_house = House("1234 Elm Street")
my_house.add_room(living_room)
my_house.add_room(kitchen)

print(my_house)
```

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def display_info(self):
        return f"Student Name: {self.name}, Student ID: {self.student_id}"


class Instructor:
    def __init__(self, name, instructor_id):
        self.name = name
        self.instructor_id = instructor_id

    def display_info(self):
        return f"Instructor Name: {self.name}, Instructor ID: {self.instructor_id}"


class CourseMaterial:
    def __init__(self, material_name):
        self.material_name = material_name

    def display_info(self):
        return f"Course Material: {self.material_name}"


class UniversityCourse:
    def __init__(self, course_name, students, instructors, course_materials):
        self.course_name = course_name
        self.students = students
        self.instructors = instructors
        self.course_materials = course_materials

    def display_info(self):
        print(f"Course Name: {self.course_name}")
        print("Students:")
        for student in self.students:
            print(student.display_info())
        print("Instructors:")
        for instructor in self.instructors:
            print(instructor.display_info())
        print("Course Materials:")
        for material in self.course_materials:
            print(material.display_info())


# Example usage:
student1 = Student("Alice", "S001")
student2 = Student("Bob", "S002")
instructor1 = Instructor("John", "I001")
instructor2 = Instructor("Jane", "I002")
material1 = CourseMaterial("Lecture Slides")
material2 = CourseMaterial("Textbook")

students = [student1, student2]
instructors = [instructor1, instructor2]
course_materials = [material1, material2]

course = UniversityCourse("Introduction to Python", students, instructors, course_materials)
course.display_info()
```

```python
class Ingredient:
    def __init__(self, name: str, quantity: float, unit: str):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def __str__(self):
        return f"{self.quantity} {self.unit} of {self.name}"

class Dish:
    def __init__(self, name: str, ingredients: list):
        self.name = name
        self.ingredients = ingredients  # List of Ingredient objects

    def __str__(self):
        ingredients_list = ", ".join(str(ingredient) for ingredient in self.ingredients)
        return f"{self.name}: {ingredients_list}"

class Menu:
    def __init__(self, name: str, dishes: list):
        self.name = name
        self.dishes = dishes  # List of Dish objects

    def __str__(self):
        dishes_list = "\n".join(str(dish) for dish in self.dishes)
        return f"Menu: {self.name}\n{dishes_list}"

class Restaurant:
    def __init__(self, name: str, menus: list):
        self.name = name
        self.menus = menus  # List of Menu objects

    def __str__(self):
        menus_list = "\n".join(str(menu) for menu in self.menus)
        return f"Restaurant: {self.name}\n{menus_list}"

# Example usage

# Creating ingredients
tomato = Ingredient("Tomato", 2, "pieces")
lettuce = Ingredient("Lettuce", 1, "head")
cheese = Ingredient("Cheese", 100, "grams")

# Creating dishes
salad = Dish("Salad", [lettuce, tomato])
cheese_pizza = Dish("Cheese Pizza", [tomato, cheese])

# Creating menus
lunch_menu = Menu("Lunch Menu", [salad])
dinner_menu = Menu("Dinner Menu", [cheese_pizza])

# Creating a restaurant
my_restaurant = Restaurant("My Restaurant", [lunch_menu, dinner_menu])

print(my_restaurant)
```

```python
class Comment:
    def __init__(self, content: str, author: 'User'):
        self.content = content
        self.author = author  # Reference to a User object

    def __str__(self):
        return f"Comment by {self.author.name}: {self.content}"

class Post:
    def __init__(self, title: str, content: str, author: 'User'):
        self.title = title
        self.content = content
        self.author = author  # Reference to a User object
        self.comments = []  # List of Comment objects

    def add_comment(self, comment: Comment):
        self.comments.append(comment)

    def __str__(self):
        comments_str = "\n  ".join(str(comment) for comment in self.comments)
        return f"Post: {self.title}\nContent: {self.content}\nAuthor: {self.author.name}\nComments:\n  {comments_str}"

class User:
    def __init__(self, name: str):
        self.name = name
        self.posts = []  

    def add_post(self, post: Post):
        self.posts.append(post)

    def __str__(self):
        posts_str = "\n".join(str(post) for post in self.posts)
        return f"User: {self.name}\nPosts:\n{posts_str}"

# Example usage

# Creating users
alice = User("Alice")
bob = User("Bob")

# Creating posts
post1 = Post("My First Post", "This is the content of my first post.", alice)
post2 = Post("A Day in the Life", "Today, I went to the park.", bob)

# Adding posts to users
alice.add_post(post1)
bob.add_post(post2)

# Creating comments
comment1 = Comment("Great post!", bob)
comment2 = Comment("Thanks for sharing!", alice)

# Adding comments to posts
post1.add_comment(comment1)
post2.add_comment(comment2)

# Display user information
print(alice)
print("\n")
print(bob)
```
